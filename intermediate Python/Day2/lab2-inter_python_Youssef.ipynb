
## Q1: Draw a line in a diagram from position (1,2) to position (10,20)

 - 2 using marker type  "star" to represent the points and marker color :red|

##  Q2:   get the local min of the function below using scipy. then plot the equation with the local min point using matlibplot


```python
from scipy.optimize import root,minimize
import matplotlib.pyplot as plt

def equ(x):
    return (x**2) + (x) + 2

optim_root= root(equ,0)

optim_root.x
```

```python
optim_min=minimize(equ,2)
```

```python
optim_min
```

```python
x=np.linspace(-2,7,2000)
plt.plot(x,equ(x))
plt.plot(optim_root.x,equ(optim_root.x),"dr",ms=10)
plt.plot(optim_min.x,equ(optim_min.x),"ok",ms=10)
plt.axhline(0,lw=0.2,color="gray")
```

## Q3 compute the solution of the system of equations using scipy




```python
#importing the scipy and numpy packages
from scipy import linalg
import numpy as np

#Declaring the numpy arrays
a = np.array([[2, 1], [1, 1]])
b = np.array([1, 1])

#Passing the values to the solve function
x = linalg.solve(a, b)

#printing the result array
print(x)
```

## Q4 use the best interpolation method to get Y_new  for this data 

        x = np.arange(0,360,5)
        y = np.sin(np.deg2rad(x))**2-np.cos(np.deg2rad(x))**2
        
   &    
        x_new = np.arange(0,360,10)



```python
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from scipy.interpolate import interp1d
from scipy.interpolate import Rbf

x = np.arange(0,360,5)
y = np.sin(np.deg2rad(x))**2-np.cos(np.deg2rad(x))**2
plt.plot(x, y, 'ro', ms=5)
plt.show()
```

```python
#that was so bad (overfitting)
onedinter=interp1d(x,y)
x_new = np.arange(0,360,10)
plt.plot(x_new, onedinter(x_new), 'g', x, y, 'ro', ms=5 )
plt.show()
```

```python
spl = UnivariateSpline(x, y)
plt.plot(x_new, spl(x_new), 'g', x, y, 'ro', ms=5)
plt.show()
```

```python
#I think that was the best fitting
spl.set_smoothing_factor(0)
plt.plot(x_new, spl(x_new), 'b', x, y, 'ro', ms=5)
plt.show()
```

```python
#I think that was the best fitting too
radial_fun=Rbf(x,y)
ynew=radial_fun(x_new)
plt.plot(x_new, ynew, 'b', x, y, 'ro', ms=5)
plt.show()
```
