## Q1: Draw a line in a diagram from position (1,2) to position (10,20)
 - 1 set the plot title >> "points plot" , xlabel: "x-axis" , ylabel : "y-axis"
 - 2 using marker type  "star" to represent the points and marker color :red|


```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2])
ypoints = np.array([10, 20])

plt.plot(xpoints, ypoints, '*r-')
plt.title("points plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.savefig("firstplot.png")
plt.show()
```

##  Q2:   get the local min of the function below using scipy. then plot the equation with the local min point using matlibplot


```python
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import numpy as np

def eqn(x):
  return x**2 + x + 2

mymin = minimize(eqn, 0, method='BFGS')

print(mymin.x)
```

```python
x_range= np.linspace(-2,2,1000)
```

```python
plt.plot(x_range,eqn(x_range))
plt.plot(mymin.x,eqn(mymin.x),"rd")
```

## Q3 compute the solution of the system of equations using scipy




```python
import scipy.linalg as la

A = np.array([[2,1],[1,1]])
print(f"A= {A}\n")

b = np.array([1,1])
print(f"b= {b}\n")

x = la.solve(A,b)
print(f" x,y = {x}\n")
A@x
```

###  Q4 use the best interpolation method to get Y_new  for this data 

        x = np.arange(0,360,5)
        y = np.sin(np.deg2rad(x))**2-np.cos(np.deg2rad(x))**2
        
   &    
        x_new = np.arange(0,360,10)


```python
from scipy.interpolate import UnivariateSpline
from scipy.interpolate import interp1d
from scipy.interpolate import Rbf
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,360,50)
y = np.sin(np.deg2rad(x))**2-np.cos(np.deg2rad(x))**2
x_new = np.arange(0,360,10)
splines = UnivariateSpline(x,y)
y_new=splines(x_new)
plt.plot(x,y,'o',x_new,y_new,'-x')
plt.show()
```

```python
splines.set_smoothing_factor(0)
y_new=splines(x_new)
plt.plot(x,y,'o',x_new,y_new,'-x')
plt.show()
```

```python
## best interpolation method for this case
radian=Rbf(x,y)
y_new=radian(x_new)
plt.plot(x,y,'o',x_new,y_new,'-x')
plt.show()
```

```python
## the worest solution 
intr1d=interp1d(x,y)
y_new=intr1d(x_new)
plt.plot(x,y,'o',x_new,y_new,'-x')
plt.show()
```
