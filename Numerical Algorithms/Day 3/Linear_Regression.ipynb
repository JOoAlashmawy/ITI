# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.linspace(0, 10, 100)
y = 3 + 1 * x  + np.random.normal(0, 1, size=100)

plt.figure(figsize=(10, 8))
plt.scatter(x, y, c='b', marker='x')
plt.show()
```

## Fit a line

Use the Moore-Penrose inverse to find the best line that fits the data.

```python
x = x.reshape(100,1)
A = np.hstack((x, np.ones(np.shape(x))))

A_plus = np.linalg.pinv(A)
coefs = A_plus.dot(y)

x_line = np.linspace(0, 10, 100)
y_line = coefs[0]*x + coefs[1]

plt.plot(x, y, "*")
plt.plot(x_line, y_line)
plt.show()
```

```python
m, b = np.polyfit(x, y, 1)
plt.figure(figsize=(10, 8))
plt.scatter(x, y, c='b', marker='x')

plt.plot(x, m*x + b)
plt.show()
```
