```python
n_samples = 1500

dataset1 = sklearn.datasets.make_circles(n_samples=n_samples, factor=.5, noise=.05)
dataset1 = pd.DataFrame(dataset1[0])
dataset2 = sklearn.datasets.make_moons(n_samples=n_samples, noise=.05)
dataset2=pd.DataFrame(dataset2[0])

dataset3 = sklearn.datasets.make_blobs(n_samples=n_samples, random_state=8)
dataset3=pd.DataFrame(dataset3[0])

dataset4 = np.random.rand(n_samples, 2), None
dataset4 = pd.DataFrame(dataset4[0])

dataset5 = pd.read_csv("dataset5.csv")
dataset6 = pd.read_csv("dataset6.csv")
dataset7 = pd.read_csv("dataset7.csv")
dataset8 = pd.read_csv("dataset8.csv")
dataset9 = pd.read_csv("dataset9.csv")

datasets = [dataset1, dataset2, dataset3, dataset4, dataset5, dataset6, dataset7, dataset8, dataset9]

# z = StandardScaler()
# dataset6[["x","y"]] = z.fit_transform(dataset6)
# ggplot(dataset6, aes(x = "x", y = "y")) + geom_point() 
```

```python
km = KMeans(n_clusters=2, random_state=0).fit(dataset1)
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(dataset1.x, dataset1.y, c=km.labels_, s=50)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.colorbar(scatter)
fig.show()
```

```python
aggler = AgglomerativeClustering(n_clusters=2, linkage='ward').fit(dataset1)
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(dataset1.x, dataset1.y, c=aggler.labels_, s=50)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.colorbar(scatter)
fig.show()
```

```python
dbscan = DBSCAN(eps=.3, min_samples=10).fit(dataset1)
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(dataset1.x, dataset1.y, c=dbscan.labels_, s=50)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.colorbar(scatter)
fig.show()
```

```python
clustering_labels = dbscan.fit_predict(dataset1) 
silhouette_score(dataset1, clustering_labels)
```

```python
km = KMeans(n_clusters=3, random_state=0).fit(dataset6)
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(dataset6.x, dataset6.y, c=km.labels_, s=50)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.colorbar(scatter)
fig.show()


aggler = AgglomerativeClustering(n_clusters=3, linkage='ward').fit(dataset6)
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(dataset6.x, dataset6.y, c=aggler.labels_, s=50)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.colorbar(scatter)
fig.show()


dbscan = DBSCAN(eps=.3, min_samples=5).fit(dataset6)
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(dataset6.x, dataset6.y, c=dbscan.labels_, s=50)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.colorbar(scatter)
fig.show()
```

```python
nclusters = range(2,20)
sil_scor = []
for i in nclusters:
    agg_clustering = AgglomerativeClustering(n_clusters=i, linkage='ward', affinity='euclidean')
    fit_agg = agg_clustering.fit_predict(dataset6)
    sil_scor.append(silhouette_score(dataset6, fit_agg, random_state=0))
    
sns.pointplot(list(nclusters),sil_scor)
plt.title('Silhouette Score Vs N Clusters')
plt.xlabel('N of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
clustering_labels = dbscan.fit_predict(dataset6) 
silhouette_score(dataset6, clustering_labels)
```
