```python
#!pip install pandas plotnine
# use any plotting library seaborn matplotlib,..........
from sklearn.neighbors import NearestNeighbors

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm
import pandas as pd
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from scipy import stats
from sklearn.metrics import silhouette_score
from sklearn import datasets

from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
```

```python
n_samples = 1500
dataset1 = datasets.make_circles(n_samples=n_samples, factor=.5, noise=.05)
dataset1=pd.DataFrame(dataset1[0],columns=['x','y'])
dataset2 = datasets.make_moons(n_samples=n_samples, noise=.05)
dataset2=pd.DataFrame(dataset2[0],columns=['x','y'])

dataset3 = datasets.make_blobs(n_samples=n_samples, random_state=8)
dataset3=pd.DataFrame(dataset3[0],columns=['x','y'])

dataset4= np.random.rand(n_samples, 2), None
dataset4=pd.DataFrame(dataset4[0],columns=['x','y'])

dataset5= pd.read_csv("data/dataset5.csv",names=['x','y'], skiprows=[0])
dataset6= pd.read_csv("data/dataset6.csv",names=['x','y'],skiprows=[0])
dataset7= pd.read_csv("data/dataset7.csv",names=['x','y'],skiprows=[0])
dataset8= pd.read_csv("data/dataset8.csv",names=['x','y'],skiprows=[0])
dataset9= pd.read_csv("data/dataset9.csv",names=['x','y'],skiprows=[0])
```

1) plot all dataset
2) use standard scalar
3) use K-means, agglomerative, dbscan
4) plot dataset after clustering
5) Evaluate inertia,silhouette score,..........

```python
def plot_data_scatter(data,hue=None,title=''):
    sns.scatterplot(data=data,hue=hue,x='x',y='y')
    plt.title(title)
    plt.show()
```

```python
def standerdize(data):
    z = StandardScaler()
    data[["x","y"]] = z.fit_transform(data)
```

---
# Dataset1

```python
plot_data_scatter(dataset1,title='Original')
```

```python
standerdize(dataset1)
```

```python
kmeans = KMeans(n_clusters=2,random_state = 42)
fit_kmeans = kmeans.fit(dataset1)
d1_labels = fit_kmeans.labels_
print('KMeans silhouette score : ',silhouette_score(dataset1,d1_labels,random_state=0)) 
```

```python
plot_data_scatter(dataset1,hue =fit_kmeans.labels_,title='Kmeans Labels')
```

```python
agg_clustering = AgglomerativeClustering(n_clusters=2,linkage='ward',affinity='euclidean')
fit_agg = agg_clustering.fit_predict(dataset1)
print('Agglomerative silhouette score : ',silhouette_score(dataset1,fit_agg,random_state=0)) 
```

```python
plot_data_scatter(dataset1,hue =fit_agg,title='Agglomerative Labels')
```

```python
db_clustering = DBSCAN(eps=0.2)
fit_db = db_clustering.fit(dataset1)
print('DBSCAN silhouette score : ',silhouette_score(dataset1,fit_db.labels_,random_state=0)) 
```

```python
plot_data_scatter(dataset1,hue =fit_db.labels_,title='DBScan Labels')
```

```python
#--- dbscan is prefered
```
