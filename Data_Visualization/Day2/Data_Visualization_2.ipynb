```python
app=JupyterDash()

app.layout = html.Div(children = [

    html.H1("My Dashboard", className="h1"),

    html.Div([
            html.Div(dcc.Graph(id='graph')),
            dcc.Slider(
                id='slider',
                min=df.year.min(),
                max=df.year.max(),
                marks={i: '{}'.format(i) for i in df.year.unique().tolist()},
                step=None,
    ),
            dcc.Dropdown(
                id = 'dropdown',
                options=[
                    {'label': str(i), 'value': str(i)} for i in df['continent'].unique()
                    
                ],
                value=None,
                placeholder='choose a continent ...',
            )
    ])
], className='row')

@app.callback(
    Output('graph', 'figure'),
    Input('slider', 'value'),
    Input('dropdown', 'value')
)
def update_fig(sliderVal, dropdownVal):
    if dropdownVal == None :
        filtered_df = df[(df.year==sliderVal)]
        fig=px.scatter(filtered_df,x='gdpPercap',
           y='lifeExp',title='World', size_max=60, color='continent',
           log_x=True,size='pop',hover_name='country')
    else:
        filtered_df = df[(df.year==sliderVal) & (df.continent==dropdownVal)]
        fig=px.scatter(filtered_df,x='gdpPercap',
           y='lifeExp',title='World', size_max=60, color='continent',
           log_x=True,size='pop',hover_name='country')
    
    
    return fig

app.run_server(debug=True)
```
